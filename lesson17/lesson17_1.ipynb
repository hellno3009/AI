```python
import numpy as np
import pandas as pd
scores = np.random.randint(50,101,size=(50,5))
scores_df = pd.DataFrame(scores,
                         columns=['國文', '英文', '數學', '地理', '歷史'],
                         index=range(1,51))
scores_df.index.name = '學號'
scores_df.columns.name = '科目'
```

```python
sum_series = scores_df.sum(axis=1)
mean_series = scores_df.mean(axis=1)
rank_series = sum_series.rank(method='min', ascending=False)
scores_df['總分'] = sum_series
scores_df['平均'] = mean_series
scores_df['排名'] = rank_series
scores_df
```

```python
scores_sorted_df = scores_df.sort_values(by='排名',ascending=True)
scores_sorted_df
```

```python
scores_sorted_df.iloc[:5]
```

```python
scores_sorted_df.iloc[-5:]
```

```python
data = {'年廣告費投入':[12.5, 15.3, 23.2, 26.4, 33.5, 34.4, 39.4, 45.2, 55.4, 60.9],
            '月均銷售額':[21.2, 23.9, 32.9, 34.1, 42.5, 43.2, 49.0, 52.8, 59.4, 63.5]}
data
```

```python
import pandas as pd
df = pd.DataFrame(data,index=[2011,2012,2013,2014,2015,2016,2017,2018,2019,2020])
df
```
