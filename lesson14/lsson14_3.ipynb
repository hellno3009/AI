```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0,5,0.1)
y1 = np.sin(2 * np.pi * t)
y2 = np.cos(2 * np.pi * t)

fig = plt.figure(figsize=(8,15))
axes1 = fig.add_subplot(2,1,1)
axes2 = fig.add_subplot(2,1,2)

axes1.plot(t,y1,"bo-.")
axes2.plot(t,y2,"ro-.")
```

```python
index = ["apple","banana","orange","tomato","guava"]
values = np.array([5,7,3,4,6]) * 1000
fig = plt.figure(figsize=(8,5))
axes = fig.add_subplot(1,1,1)
axes.bar(index,values)
plt.show()
```

```python
import random
with open("names.txt",encoding="utf-8") as file:
    names_str = file.read()
    display(names_str)
names_list = names_str.split(sep="\n")
names = random.choices(names_list,k=50)
```

```python
import numpy as np 
import pandas as pd

scores = np.random.randint(50,101,size=(50,5))
df1 = pd.DataFrame(scores,columns=["","","","",""],index=names)
df1
```

```python
df1.to_numpy()

df1.to_csv("students.csv")

df1.to_excel("students.xlsx",sheet_name="")
```
